```python
# Padding test
from cnn import max_pooling, forward, zero_pad, conv_backward
import numpy as np

w_test_array = np.random.randn(1, 10, 10, 3)
b_test_array = np.random.randn(8)

w_pad = zero_pad(w_test_array, 1)

print(w_pad)
```

```python
# Forward prop test
A_prev = np.random.randn(10, 32, 32, 3)
W = np.random.randn(5, 5, 3, 8)
b = np.random.randn(1, 1, 1, 8)
hparameters = {'stride' : 2, 'pad' : 2}

Z, cache = forward(A_prev, W, b, hparameters)

print(Z.shape)
print(type(cache))
print(len(cache))

```

```python
# Max pooling test
pool_test_array = np.random.randn(10, 5, 5, 2)
hparameters = {'stride' : 1, 'f' : 3}
pool_result, cache = max_pooling(pool_test_array, hparameters)

print(pool_result.shape)
print(type(cache))
print(len(cache))
```

```python
# Backpropagation in conv layer test
conv_backprop_test = np.random.randn(3, 10, 10, 7)

dA_prev, dW, db = conv_backward(conv_backprop_test, cache)
print(dA_prev.shape)
print(dW.shape)
print(db.shape)
```
